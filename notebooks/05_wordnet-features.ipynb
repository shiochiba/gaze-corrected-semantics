```python
import json
from nltk.corpus import wordnet as wn
with open("label_map.json") as l:
    label_map = json.load(l)
```

```python
unique_corrected = set(label_map.values())
print(f"Unique corrected labels: {len(unique_corrected)} (vs {len(label_map)} raw labels)")

wordnet_dict = {}
no_words = []

for corrected_label in unique_corrected:
    lemmas = wn.lemmas(corrected_label)
    if lemmas:
        wordnet_dict[corrected_label] = lemmas[0].synset()
        print(f"{corrected_label} -> {lemmas[0]}")
    else:
        no_words.append(corrected_label)
        print(f"{corrected_label} -> No synset found")

print(f"\nMatched: {len(wordnet_dict)}")
print(f"Unmatched: {len(no_words)}")
```

```python
print(len(no_words))
for w in no_words:
    print(w)
```

```python
all_lemmas = [(lemma.lower(), synset) 
                    for synset in wn.all_synsets(wn.NOUN) 
                    for lemma in synset.lemma_names()]

def suggesting_partial(substring):
    substring = substring.lower().replace(' ', '_')
    matches = {synset for lemma, synset in all_lemmas if substring in lemma}
    if not matches:
        print(f"No synsets found for '{substring}'")
        return
    for s in matches:
        print(f"{s.name()}  —  {s.definition()}")

def suggesting(word):
    synsets = wn.synsets(word, pos=wn.NOUN)
    if not synsets:
        print(f"No synsets found for '{word}'")
        return
    for s in synsets:
        print(f"{s.name()}  —  {s.definition()}")
```
